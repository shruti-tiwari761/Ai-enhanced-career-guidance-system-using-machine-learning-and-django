```python
def classify_student(row):
    # Assign weights to key responses for classification
    scores = {
        "Not Confident": 1, "Slightly Confident": 2, "Neutral/Somewhat Confident": 3, 
        "Confident": 4, "Very Confident": 5,
        "Poor": 1, "Below Average": 2, "Average": 3, "Good": 4, "Excellent": 5,
        "Not Interested": 1, "Slightly Interested": 2, "Neutral/Somewhat Interested": 3,
        "Interested": 4, "Very Interested": 5,
        "Not Important": 1, "Somewhat Important": 3, "Very Important": 5,
        "Not Creative": 1, "Slightly Creative": 2, "Neutral/Somewhat Creative": 3,
        "Creative": 4, "Very Creative": 5,
        "No": 1, "Sometimes": 2, "Yes": 3,
        "I prefer working independently": 1, "It depends on the project": 2, "I prefer working with others": 3,
        "Not Comfortable": 1, "Slightly Comfortable": 2, "Neutral/Somewhat Comfortable": 3,
        "Comfortable": 4, "Very Comfortable": 5
    }
    
    # Map scores for each column
    confidence_score = scores.get(row["How confident are you in your understanding of subjects related to Arts (e.g., History, Political Science, Literature, Geography)?  "], 0)
    performance_score = scores.get(row["How would you rate your performance in subjects related to Arts (e.g., History, Political Science, Literature, Geography)?  "], 0)
    career_interest_score = scores.get(row["How interested are you in pursuing a career in the creative or social sciences fields (e.g., Writer, Psychologist, Artist, Historian, Journalist)?  "], 0)
    importance_score = scores.get(row["How important is a career in the creative arts, humanities, or social sciences for you?  "], 0)
    creativity_score = scores.get(row["How do you rate your creative thinking or imagination?  "], 0)
    project_enjoyment_score = scores.get(row["Do you enjoy working on creative projects or activities  "], 0)
    collaboration_score = scores.get(row["Do you prefer working independently on creative projects, or do you enjoy collaborating with others?  "], 0)
    public_speaking_score = scores.get(row["How comfortable are you with expressing your thoughts or ideas publicly (e.g., speaking in front of an audience, performing on stage)?  "], 0)

    # Compute total score
    total_score = (
        confidence_score
        + performance_score
        + career_interest_score
        + importance_score
        + creativity_score
        + project_enjoyment_score
        + collaboration_score
        + public_speaking_score
    )
    
    # Classification thresholds and degree suggestions
    if total_score >= 30:
        recommendation = "Highly Recommended"
        degree = "Fine Arts, Psychology, Journalism, Creative Writing, or Literature"
    elif total_score >= 20:
        recommendation = "Recommended"
        degree = "History, Sociology, Political Science, Cultural Studies, or Philosophy"
    else:
        recommendation = "Not Recommended"
        degree = "Consider exploring alternative fields, such as STEM or Commerce"

    return recommendation, degree

# Apply the function to classify students and suggest degrees
df[["Career Guidance Recommendation", "Suggested Degree"]] = df.apply(
    lambda row: pd.Series(classify_student(row)), axis=1
)
```

```python
from sklearn.ensemble import BaggingClassifier
from sklearn.tree import DecisionTreeClassifier
```

```python
# Initialize base estimator
base_estimator = DecisionTreeClassifier()

# Define Bagging Classifier (use 'estimator' instead of 'base_estimator')
bagging_clf = BaggingClassifier(estimator=base_estimator, random_state=42)
```

```python
y_pred = bagging_clf.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score, f1_score, recall_score, precision_score
import pandas as pd

# Assuming y_test and y_pred are already defined
# y_pred is the predicted values for the test set, y_test is the actual values

print("For BaggingClassifier Model")
print("-" * 50)

# Get the number of unique classes in the target variable
n_classes = len(set(y_test))

# Confusion Matrix
cf = confusion_matrix(y_test, y_pred)

# Convert the confusion matrix to a DataFrame
cf_df = pd.DataFrame(cf, columns=[f"Class {i}" for i in range(n_classes)], index=[f"Class {i}" for i in range(n_classes)])

# Print the confusion matrix
print(cf_df)
print("-" * 50)

# Test dataset evaluation metrics
print(f"Test dataset accuracy_score: {accuracy_score(y_test, y_pred)*100:.4f}")
print(f"Test dataset f1_score: {f1_score(y_test, y_pred, average='weighted')*100:.4f}")
print(f"Test dataset recall_score: {recall_score(y_test, y_pred, average='weighted')*100:.4f}")
print(f"Test dataset precision_score: {precision_score(y_test, y_pred, average='weighted')*100:.4f}")
```
